```python
def plot_item_cnt(df):
    days_items_sum = {day : 0 for day in list(range(34))}
    for id, row in df.iterrows():
        day = row['date_block_num']
        cnt = row['item_cnt_day']
        days_items_sum[day] += cnt
    plt.plot(list(range(34)), list(days_items_sum.values()))
    plt.xlabel('date block number')
    plt.ylabel('count of sold items')
    plt.show()
```

* multivariate (vector autoregression - stats in python) vs univariate models (prophet or arima) depending on numbers of variables we are predicting 
* in machine learning we can use any kind of regressor model 
* question, how many days ahead we want to predcit
* we can train machine learning model to predict few days ahead 
* while in traditional models we forecast the next day and based on this prediction we made another prediction 



```python
dfWeekly = daysSum31['item_cnt_sum'].rolling(7).sum()
list(dfWeekly)
```

```python
dfDaybefore = daysSum31['item_cnt_sum'].rolling(2).sum() - daysSum31['item_cnt_sum']
dfDaybefore
```

```python
daysSum31['week_sum'] = dfWeekly
daysSum31['day_before'] = dfDaybefore
daysSum31 = daysSum31[6:]
daysSum31
```

```python
features = ['day_no', 'day_of_the_week', 'week_sum']
label = 'item_cnt_sum'
test_df = daysSum31[daysSum31['days'] >= pd.Timestamp('2015-05-01')]
train_df = daysSum31[daysSum31['days'] < pd.Timestamp('2015-05-01')]
X_train, y_train = train_df[features], train_df[label]
X_test, y_test = test_df[features], test_df[label]
```

```python
from xgboost import XGBRegressor
reg = XGBRegressor(n_estimators=500, learning_rate=0.01)
reg.fit(X_train, 
        y_train,
        eval_set=[(X_train, y_train), (X_test, y_test)],
        eval_metric='mae')
```

```python
predictions = reg.predict(X_test)
```

```python
test_df = test_df.reset_index().drop('index', axis=1)
test_df['predictions'] = pd.Series(predictions)
```

```python
import seaborn as sns
plt.rcParams.update({'figure.figsize': (20, 3), 'figure.dpi':300})
fig, ax = plt.subplots()
sns.lineplot(data=daysSum31.tail(300), x='days', y='item_cnt_sum')
sns.lineplot(data=test_df, x='days', y='predictions')
plt.grid(linestyle='-', linewidth=0.3)
ax.tick_params(axis='x', rotation=90)
```
